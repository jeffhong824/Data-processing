#### Q2: Create a new column named **BMI**

**HINT**: https://zh.wikipedia.org/wiki/%E8%BA%AB%E9%AB%98%E9%AB%94%E9%87%8D%E6%8C%87%E6%95%B8


```python
BMI = []
for each_column in range( len(df) ) :
    bmi = df.loc[each_column]['weight'] / ( (df.loc[each_column]['height']*0.01)**2 )
    BMI.append(bmi)
    
df['BMI']=BMI
```

```python
df['BMI2'] = df['weight'] / (df['height'] / 100.) ** 2
```
