```python
from docx import Document
import os
from win32com import client as wc
from win32com.client import Dispatch, constants
import win32com
from docx.shared import Inches
from docx.shared import RGBColor
from docx.enum.text import WD_COLOR_INDEX
```

```python
def GenerateNewWord(filename): # Python-docx 新建文檔
    document = Document()
    document.save(filename)

def TransDocToDocx(oldDocName,newDocxName): # win32com 將 doc 轉為 docx

    # 打開word應用程式
    word = wc.Dispatch('Word.Application')
    # 打開 舊word 文件
    doc = word.Documents.Open(oldDocName)
    doc.SaveAs(newDocxName, 12,Encoding='utf-8') # 參數 12 表示的是docx文件
    doc.Close
    word.Quit
    
def funOpenNewFile(filename): # 打開新的word文檔並添加內容

    # 打開word應用程式
    word = wc.Dispatch('Word.Application') # # word = DispatchEx('Word.Application')
    
    # 如果不聲明以下屬性，運行的時候會顯示的打開word
    word.Visible = 1 # 0:後台運行 1:前台運行(可見)
    word.DisplayAlerts = 0 # 不顯示，不警告

    # 創建新的word文檔
    doc = word.Documents.Add()

    # 在文檔開頭添加內容
    myRange1 = doc.Range(0, 0)
    myRange1.InsertBefore('Hello word\n')

    # 在文檔末尾添加內容
    myRange2 = doc.Range()
    myRange2.InsertAfter('Bye word\n')

    # 在文檔i指定位置添加內容
    i = 0
    myRange3 = doc.Range(0, i)
    myRange3.InsertAfter("what's up, bro?\n")

    doc.SaveAs(filename, 12,Encoding='utf-8')
    doc.Close
    word.Quit

def funOpenExistFile(oldDocName,newDocxName):

    # 打開word應用程式
    word = wc.Dispatch('Word.Application') # # word = DispatchEx('Word.Application')
    
    # 如果不聲明以下屬性，運行的時候會顯示的打開word
    word.Visible = 1 # 0:後台運行 1:前台運行(可見)
    word.DisplayAlerts = 0 # 不顯示，不警告
    
    doc = word.Documents.Open(oldDocName)
    
    # 在文檔開頭添加內容
    myRange1 = doc.Range(0, 0)
    myRange1.InsertBefore('Hello word\n')

     # 在文檔末尾添加內容
    myRange2 = doc.Range()
    myRange2.InsertAfter('Bye word\n')

    # 在文檔i指定位置添加內容
    i = 0
    myRange3 = doc.Range(0, i)
    myRange3.InsertAfter("what's up, bro?\n")

    doc.SaveAs(newDocxName, 12,Encoding='utf-8')
    doc.Close
    word.Quit
    
def funGeneratePDF(oldDocName,docx2type): # 轉換word為pdf
    
    # 打開word應用程式
    word = wc.Dispatch('Word.Application') # # word = DispatchEx('Word.Application')
    
    # 如果不聲明以下屬性，運行的時候會顯示的打開word
    word.Visible = 1 # 0:後台運行 1:前台運行(可見)
    word.DisplayAlerts = 0 # 不顯示，不警告
      
    # 創建新的word文檔
    doc = word.Documents.Add()

    # 在文檔開頭添加內容
    #myRange1 = doc.Range(0, 0)
    #myRange1.InsertBefore('Hello word\n')

    # 在文檔末尾添加內容
    #myRange2 = doc.Range()
    #myRange2.InsertAfter('Bye word\n')
        
    #打開舊word 文件
    doc = word.Documents.Open(oldDocName)
    
    # 存擋類型 # txt=4, html=10, docx=16， pdf=17 # https://blog.csdn.net/ynyn2013/article/details/49120731
    if int(docx2type) == 4 :
        newDocName = oldDocName.replace('.docx','.txt')
    elif int(docx2type) == 10 :
        newDocName = oldDocName.replace('.docx','.html')
    elif int(docx2type) == 17 :
        newDocName = oldDocName.replace('.docx','.pdf')
    else:
        newDocName = oldDocName
    doc.SaveAs(newDocName, docx2type,Encoding='utf-8')
    
    
def funwordtext(filename,pic=None):
    document = Document()
    
    # 標題
    document.add_heading('Document Title', 0)

    # 添加有粗、斜體的段落
    p = document.add_paragraph('A plain paragraph having some ')
    p.add_run('bold').bold = True # 粗體
    p.add_run(' and some ')
    p.add_run('italic ').italic = True # 斜體
    C = p.add_run('QAQ.') # 粗體 + 斜體
    C.bold = True
    C.italic = True
    D = p.add_run('~~~.') # 粗體 + 斜體
    font = D.font
    font.color.rgb = RGBColor(0x42, 0x24, 0xE9)
    
    paragraph = document.add_paragraph()
    run = paragraph.add_run('Red ')
    run.font.color.rgb = RGBColor(255, 0, 0)
    run = paragraph.add_run('Green ')
    run.font.color.rgb = RGBColor(0, 255, 0)
    run = paragraph.add_run('Blue')
    run.font.color.rgb = RGBColor(0, 0, 255)

    paragraph2 = document.add_paragraph('Highlight_')
    E = paragraph2.add_run('color')
    E.font.highlight_color = WD_COLOR_INDEX.YELLOW
    
    document.add_heading('Heading, level 1', level=1)
    document.add_paragraph('Intense quote', style='Intense Quote')
    document.add_paragraph('first item in unordered list', style='List Bullet')
    document.add_paragraph('first item in ordered list', style='List Number')
    
    # 圖片
    if str(pic) != 'None' :
        document.add_picture(pic, width=Inches(1.25))

    ## 表格
    # 待添加至表格的列表
    records = (
    (3, '101', 'Spam'),
    (7, '422', 'Eggs'),
    (4, '631', 'Spam, spam, eggs, and spam')
    )
    
    #新建表格 1行3列
    table = document.add_table(rows=1, cols=3, style="Medium Grid 1 Accent 1") # Normal Table #  Table Grid
    '''
    # type
    from docx.enum.style import WD_STYLE_TYPE
    from docx import Document

    styles = document.styles

    # 生成所有表樣式
    for s in styles:
        if s.type == WD_STYLE_TYPE.TABLE:
            print(s)
    '''
    
    # 添加標題
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Qty'
    hdr_cells[1].text = 'Id'
    hdr_cells[2].text = 'Desc'

    # 添加表格行數與填充內容
    for qty, id_, desc in records:
        row_cells = table.add_row().cells
        row_cells[0].text = str(qty)
        row_cells[1].text = id_
        row_cells[2].text = desc
    
    #document.add_page_break()

    document.save(filename)
```
