```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Raw image

Cx_ray = cv2.imread('Xray.jpg')
Cx_ray = Cx_ray[:,:,::-1]
plt.imshow(Cx_ray)
```

```python
# Gray Image

Cx_ray_gray = cv2.cvtColor(Cx_ray, cv2.COLOR_RGB2GRAY)
plt.imshow(Cx_ray_gray, cmap='gray')
```

```python
# Binary Thresholding

ret, img_th = cv2.threshold(Cx_ray_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img_th, cmap='gray')
```

```python
# Blur

img_blurred = cv2.GaussianBlur(Cx_ray_gray, (11, 11), 0)
plt.imshow(img_blurred, cmap='gray')
```

```python
# Sharpening

kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
img_sharpened = cv2.filter2D(Cx_ray_gray, -1, kernel)
plt.imshow(img_sharpened, cmap='gray')
```

```python
# Sobel Edge Detection

# Edge Detection - Sobel
sobelX = cv2.Sobel(Cx_ray_gray, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(Cx_ray_gray, cv2.CV_64F, 0,1)
# absolute value
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
sobel_com = cv2.bitwise_or(sobelX, sobelY)

plt.imshow(sobelX, cmap='gray')
plt.show()
plt.imshow(sobelY, cmap='gray')
plt.show()
plt.imshow(sobel_com, cmap='gray')
plt.show()
```

```python
# Edge Detection - Canny
img_canny = cv2.Canny(Cx_ray_gray, 30, 130)
plt.imshow(img_canny, cmap='gray')
```
